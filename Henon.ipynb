```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def henon_map(x, y, a=1.4, b=0.3):
    x_new = 1 - a * x ** 2 + y
    y_new = b * x
    return x_new, y_new

# Parameters
a = 1.4
b = 0.3
num_iterations = 100000

# Initial conditions
x_values = np.zeros(num_iterations)
y_values = np.zeros(num_iterations)

# Starting point
x_values[0], y_values[0] = (0, 0)

# Generate the Hénon attractor
for i in range(1, num_iterations):
    x_values[i], y_values[i] = henon_map(x_values[i-1], y_values[i-1], a, b)


# Plot the Hénon attractor
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, marker='o', linestyle = '', color='black', markersize=0.5)
plt.title('Hénon Attractor')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
